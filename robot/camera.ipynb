```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
data = np.load("./images/capture_20250620_163432.npz")
color_image = data["color"]
depth_image = data["depth"]
```

```python
depth_colormap = cv2.normalize(depth_image, None, 0, 255, cv2.NORM_MINMAX)
depth_colormap = np.uint8(depth_colormap)
depth_colormap = cv2.cvtColor(depth_colormap, cv2.COLOR_GRAY2BGR)  # (480, 640, 3)

# Combine side-by-side
combined_image = np.hstack((color_image, depth_colormap))

# Convert BGR (OpenCV) to RGB (matplotlib)
combined_rgb = cv2.cvtColor(combined_image, cv2.COLOR_BGR2RGB)

# Display
plt.figure(figsize=(12, 6))
plt.imshow(combined_rgb)
plt.axis('off')
plt.title("Color + Depth Side-by-Side")
plt.show()
```
